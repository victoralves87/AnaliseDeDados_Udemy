```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
import statistics  as sts
```

```python
#importar dados"
dataset = pd.read_csv("/home/victor87/Área de trabalho/ciência de dados/FormacaoCD/todos os arquivos/13.Prática em Python/dados/dados.csv", sep=";")
dataset.head()
```

```python
dataset.shape
```

```python
#contamos valores NAN
dataset.isnull().sum()
```

```python
#Agrupando os municipios pelo PIB
agrupado = dataset.groupby("MUNICIPIO")["PIB"].sum()
agrupado
```

```python
# Agrupar os municípios pelo PIB e garantir que o PIB seja numérico
agrupado = dataset.groupby("MUNICIPIO")["PIB"].sum()

# Garantir que o PIB seja numérico
agrupado = pd.to_numeric(agrupado, errors='coerce')

# Selecionar os 10 maiores PIBs
top_10_pib = agrupado.sort_values(ascending=False).head(10)

# Criar o gráfico de barras para os 10 maiores PIBs
plt.figure(figsize=(10, 6))
top_10_pib.plot(kind='bar', color='teal')

# Adicionar título e rótulos
plt.title('Top 10 Maiores PIBs por Município', fontsize=13)
plt.xlabel('Municípios', fontsize=14)
plt.ylabel('PIB Total', fontsize=14)
plt.xticks(rotation=80, ha='right', fontsize=10)  # o ângulo dos rótulos dos municípios

# Exibir o gráfico
plt.show()
```

```python
#Agrupando empenho em diárias em municipios gaúchos.
agrupado2 = dataset.groupby("MUNICIPIO")["VALOREMPENHO"].sum()
agrupado2
```

```python
# Agrupar os municípios pelo valor de empenho e garantir que o valor de empenho seja numérico
agrupado2 = dataset.groupby("MUNICIPIO")["VALOREMPENHO"].sum()

# Garantir que o valor de empenho seja numérico
agrupado2 = pd.to_numeric(agrupado2, errors='coerce')

# Selecionar os 10 maiores valores de empenho
top_10_empenho = agrupado2.sort_values(ascending=False).head(10)

# Criar o gráfico de barras para os 10 maiores valores de empenho
plt.figure(figsize=(10, 6))
top_10_empenho.plot(kind='bar', color='teal')

# Adicionar título e rótulos
plt.title('Top 10 Maior Valor de Empenho por Município', fontsize=16)
plt.xlabel('Municípios', fontsize=14)
plt.ylabel('Valor de Empenho', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Ajuste o ângulo dos rótulos dos municípios

# Exibir o gráfico
plt.show()
```
